```python
!pip install tensorflow opencv-python-headless numpy matplotlib

import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
import pandas as pd

(X_train_mnist, y_train_mnist), (X_test_mnist, y_test_mnist) = mnist.load_data()

kaggle_az_path = '/content/A_Z Handwritten Data.csv'
kaggle_az_data = pd.read_csv(kaggle_az_path)

X_kaggle = kaggle_az_data.drop('0', axis=1).values
y_kaggle = kaggle_az_data['0'].values

X_kaggle = X_kaggle.reshape(-1, 28, 28)

X_data = np.concatenate([X_train_mnist, X_kaggle], axis=0)
y_data = np.concatenate([y_train_mnist, y_kaggle + 10], axis=0)

X_data = X_data / 255.0
X_data = X_data.reshape(-1, 28, 28, 1)

plt.imshow(X_data[0].reshape(28, 28), cmap='gray')
plt.title(f"Sample Label: {y_data[0]}")
plt.show()

num_classes = len(np.unique(y_data))
y_data = tf.keras.utils.to_categorical(y_data, num_classes=num_classes)

from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_data, y_data, test_size=0.2, random_state=42)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

print("Project updated with MNIST and Kaggle A-Z datasets. Model training completed.")
```

```python
def test_custom_image(image_path):
    """Load and preprocess a custom image, then make a prediction with better image handling."""

    test_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if test_img is None:
        print("Error: Unable to load image. Check the file path.")
        return

    _, test_img = cv2.threshold(test_img, 128, 255, cv2.THRESH_BINARY_INV)
    test_img = cv2.resize(test_img, (28, 28), interpolation=cv2.INTER_AREA)

    test_img = test_img / 255.0
    test_img = test_img.reshape(1, 28, 28, 1)

    prediction = model.predict(test_img)
    predicted_class = np.argmax(prediction)


    if predicted_class < 10:
        readable_char = str(predicted_class)
    else:
        readable_char = chr(predicted_class + 55)

    print(f"Predicted character: {readable_char}")
    plt.imshow(test_img.reshape(28, 28), cmap='gray')
    plt.title(f"Predicted Character: {readable_char}")
    plt.show()

image_path = '/content/happy_new_year_2025.png'
test_custom_image(image_path)
```

```python
!apt-get install tesseract-ocr
!pip install pytesseract

import cv2
import pytesseract
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

image_path = '/content/happy_new_year_2025.png'

img = cv2.imread(image_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blurred = cv2.GaussianBlur(gray, (5, 5), 0)

_, thresh = cv2.threshold(blurred, 128, 255, cv2.THRESH_BINARY_INV)

kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
morph_img = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

custom_config = r'--oem 3 --psm 6'

text = pytesseract.image_to_string(morph_img, config=custom_config)

print("Распознанный текст:", text)

plt.imshow(morph_img, cmap='gray')
plt.title("Обработанное изображение для OCR")
plt.show()
```
